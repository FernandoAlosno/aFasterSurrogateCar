```python
import pandas as pd
import numpy as np

import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots


df = pd.read_csv('All_Samples_wing_350.csv')

identifier = 'filename'
responses = ['Cd', 'Cl', 'CmPitch', 'L_D_Ratio']
features = [col for col in df.columns if col not in responses and col != identifier]
features = features[:-1]
# print(df.columns)
print(len(features))
print(features)
```

```python
df_nonan = df.dropna()
df_reached = df_nonan[df_nonan['Reached_500']]
df_reached = df_reached.drop('Reached_500', axis=1, inplace=False)
df[df['Reached_500']==False]
```

```python
for response in responses:
    fig = px.histogram(df_reached, x=response, 
                       title=f'Distribution of Response: {response}', 
                       marginal='box', # Adds a boxplot above the histogram
                       template='plotly_white')
    fig.show()
```

```python
# Original L_D_Ratio -4.70
# Una fracción considerable sampleada por encima del original, la mayoria por debajo
```

```python
from itertools import combinations

df_extend = df_reached.copy()
nuevas_features = []

puntos_base = list(set([col.rsplit('_', 1)[0] for col in features if col.endswith(('_dx', '_dy', '_dz'))]))

pares_puntos = list(combinations(puntos_base, 2))

for p1, p2 in pares_puntos:
    if all(f"{p}_{eje}" in df.columns for p in [p1, p2] for eje in ['dx', 'dy', 'dz']):
        nombre_col_dist = f"dist_{p1}_to_{p2}"
        nombre_col_angl = f"angl_{p1}_to_{p2}"
        
        dx_diff = df[f"{p1}_dx"] - df[f"{p2}_dx"]
        dy_diff = df[f"{p1}_dy"] - df[f"{p2}_dy"]
        dz_diff = df[f"{p1}_dz"] - df[f"{p2}_dz"]

        norm_p1 = np.sqrt(df[f"{p1}_dx"]**2 + df[f"{p1}_dy"]**2 + df[f"{p1}_dz"]**2)
        norm_p2 = np.sqrt(df[f"{p2}_dx"]**2 + df[f"{p2}_dy"]**2 + df[f"{p2}_dz"]**2)
        dot_p1_p2 = df[f"{p1}_dx"]*df[f"{p2}_dx"] + df[f"{p1}_dy"]*df[f"{p2}_dy"] + df[f"{p1}_dz"]*df[f"{p2}_dz"]

        df_extend[nombre_col_dist] = np.sqrt(dx_diff**2 + dy_diff**2 + dz_diff**2)
        nuevas_features.append(nombre_col_dist)
        df_extend[nombre_col_angl] =  dot_p1_p2/norm_p1/norm_p2
        nuevas_features.append(nombre_col_angl)

print(f"Se han creado {len(nuevas_features)} nuevas variables de distancia relativa.")
df_extend.describe()
```

## Feature Ranking 

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.ensemble import RandomForestRegressor

X_completo = df_extend[features+nuevas_features]
# y_reg = df_extend['Cd']
# y_reg = df_extend['Cl']
# y_reg = df_extend['CmPitch']
y_reg = df_extend['L_D_Ratio']


rf_dist = RandomForestRegressor(n_estimators=100, random_state=42)
kf = KFold(n_splits=5, shuffle=True, random_state=42)

print("\n--- Paso 2: Evaluando Random Forest (Features Base + Distancias) ---")
scores_rf_dist = cross_val_score(rf_dist, X_completo, y_reg, cv=kf, scoring='r2')

print(f"R² Medio = {np.mean(scores_rf_dist):.3f} (+/- {np.std(scores_rf_dist):.3f})")
```

```python
# ---------------------------------------------------------
# Analizar la Importancia de las Características (Feature Importance)
# ---------------------------------------------------------
rf_dist.fit(X_completo, y_reg)
importancias = rf_dist.feature_importances_

df_imp = pd.DataFrame({
    'Feature': X_completo.columns,
    'Importancia': importancias
}).sort_values(by='Importancia', ascending=False).head(len(X_completo.columns)//5)

fig_imp = px.bar(
    df_imp, 
    x='Importancia', 
    y='Feature', 
    orientation='h',
    title='Top Características Predictivas (Incluyendo Distancias Relativas)',
    color='Importancia',
    color_continuous_scale='Viridis',
    template='plotly_white'
)
fig_imp.update_layout(yaxis={'categoryorder':'total ascending'}) 
fig_imp.show()
print("Numero de Features Relevantes", len(list(df_imp['Feature'])))
```

```python
from sklearn.model_selection import cross_validate
from sklearn.svm import SVR
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.pipeline import make_pipeline
from sklearn.neural_network import MLPRegressor
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, Matern, ConstantKernel as C
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression, Lasso, ElasticNet, HuberRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.ensemble import ExtraTreesRegressor, GradientBoostingRegressor, HistGradientBoostingRegressor

import warnings
from sklearn.exceptions import ConvergenceWarning
warnings.filterwarnings("ignore", category=ConvergenceWarning)

relevant_features_1 = list(df_imp1['Feature'])
X_relev = X_relevant_1[relevant_features_1] 

models_advanced = {
    "Linear Regression": LinearRegression(),
    # "Lasso": Lasso(alpha=0.001, max_iter=10000, random_state=42),
    # "ElasticNet": ElasticNet(alpha=0.001, l1_ratio=0.5, max_iter=10000, random_state=42),
    # "Robust Regression (Huber)": HuberRegressor(),
    
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    
    "Radial Basis (SVR RBF)": SVR(kernel='rbf', C=10.0, gamma='scale'),
    # "SVR Linear": SVR(kernel='linear', C=1.0),
    # "SVR Polynomial": SVR(kernel='poly', degree=3, C=10.0, gamma='scale'),
    
    # "Polinomial (Grado 2 + Ridge)": make_pipeline(
    #     PolynomialFeatures(degree=2, include_bias=False),
    #     Ridge(alpha=10.0, random_state=42)
    # ),
    # "Polynomial Degree 3 + Ridge": make_pipeline(
    # PolynomialFeatures(degree=3, include_bias=False),
    # Ridge(alpha=1.0)
    # ),
    # "Interaction Only Polynomial": make_pipeline(
    #     PolynomialFeatures(degree=2, interaction_only=True, include_bias=False),
    #     Ridge(alpha=1.0)
    # ),
    
    "XGBoost": XGBRegressor(
        n_estimators=500, 
        learning_rate=0.05, 
        max_depth=6, 
        subsample=0.8, 
        colsample_bytree=0.8, 
        random_state=42
    ),
    "LightGBM": LGBMRegressor(
        n_estimators=500,
        learning_rate=0.05,
        num_leaves=31,
        max_depth=6,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42,
        verbose=-1
    ),
    "AdaBoost": AdaBoostRegressor(
        estimator=DecisionTreeRegressor(max_depth=6, random_state=42),
        n_estimators=500,
        learning_rate=0.05,
        random_state=42
    ),
    # "Extra Trees": ExtraTreesRegressor(
    # n_estimators=500,
    # max_depth=None,
    # random_state=42,
    # n_jobs=-1
    # ),
    # "Gradient Boosting": GradientBoostingRegressor(
    #     n_estimators=500,
    #     learning_rate=0.05,
    #     max_depth=4,
    #     random_state=42
    # ),
    # "Histogram Gradient Boosting": HistGradientBoostingRegressor(
    #     max_iter=500,
    #     learning_rate=0.05,
    #     max_depth=6,
    #     random_state=42
    # ),
    
    "Neural Network (MLP)": MLPRegressor(
        hidden_layer_sizes=(128, 64, 32), 
        activation='relu', 
        solver='adam', 
        max_iter=2000, 
        early_stopping=True,
        random_state=42
    ),
    
    "Gaussian Process (Kriging)": GaussianProcessRegressor(
        kernel=C(1.0, (1e-3, 1e3)) * Matern(length_scale=1.0, nu=1.5),
        alpha=0.1, 
        n_restarts_optimizer=10,
        random_state=42
    )
}

kf = KFold(n_splits=5, shuffle=True, random_state=42)

print("--- Evaluando Modelos Avanzados de Regresión ---\n")

results_r2 = []

for name, model in models_advanced.items():
    cv_res = cross_validate(model, X_relev, y_reg, cv=kf, scoring='r2', return_train_score=True)
    
    train_scores = cv_res['train_score']
    test_scores = cv_res['test_score']
    
    print(f"{name:<40}")
    print(f"  > Train R²: {np.mean(train_scores):.3f} (+/- {np.std(train_scores):.3f})")
    print(f"  > Test  R²: {np.mean(test_scores):.3f} (+/- {np.std(test_scores):.3f})")
    
    for score in test_scores:
        results_r2.append({'Modelo': name+"Relev", 'R2_Score': score})

df_results = pd.DataFrame(results_r2)

fig_comp = px.box(
    df_results, 
    x='R2_Score', 
    y='Modelo', 
    color='Modelo',
    title='Comparación de Estabilidad (R² en 5-Fold CV)',
    points='all',
    template='plotly_white'
)

fig_comp.add_vline(x=0.5, line_dash="dash", line_color="red", annotation_text="Techo Anterior")
fig_comp.update_layout(xaxis_title="Score R² (Más cerca de 1.0 es mejor)", showlegend=False)

fig_comp.show()
```

```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
from scipy.stats import spearmanr

X, y = X_relev, y_reg
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=43)

models_and_parameters = {
    'RandomForest': {
        'model': RandomForestRegressor(random_state=42),
        'params': {
            'n_estimators': [50, 100, 200],
            'max_depth': [None, 10, 20],
            'min_samples_split': [2, 5]
        }
    },
    'XGBoost': {
        'model': XGBRegressor(random_state=43, objective='reg:squarederror'),
        'params': {
            'n_estimators': [50, 100, 200],
            'learning_rate': [0.01, 0.1, 0.2],
            'max_depth': [3, 5, 7]
        }
    },
        'LightGBM': {
        'model': LGBMRegressor(random_state=42, verbose=-1, n_jobs=1), 
        'params': {
            'n_estimators': [50, 100, 200],
            'learning_rate': [0.01, 0.1, 0.2],
            'num_leaves': [31, 50, 100]
        }
    },
    'AdaBoost': {
        'model': AdaBoostRegressor(
            estimator=DecisionTreeRegressor(random_state=42),
            random_state=42),
        'params': {
            'n_estimators': [100, 500],
            'learning_rate': [0.01, 0.05, 0.1],
            'loss': ['linear', 'square', 'exponential'],
            'estimator__max_depth': [4, 6, 10]
        }
    },
    "Gaussian Process (Kriging)": {
        'model': GaussianProcessRegressor(n_restarts_optimizer=5, random_state=42),
        'params':{
            'alpha': [1e-2, 0.1, 1.0],
            'kernel': [
                C(1.0) * RBF(length_scale=1.0),
                C(1.0) * Matern(length_scale=1.0, nu=1.5),
                C(1.0) * Matern(length_scale=1.0, nu=2.5)
            ]
        }
    },
    # "Radial Basis (SVR RBF)": {
    #     'model': SVR(kernel='rbf'),
    #     'params': {
    #         'C': [0.1, 1, 10, 100],
    #         'gamma': ['scale', 'auto', 0.01, 0.1],
    #         'epsilon': [0.01, 0.1, 0.2]
    #     }
    # }
    
}

results = []

print("Starting Grid Search...\n")
for name, mp in models_and_parameters.items():
    print(f"Fitting {name}...")
    
    clf = GridSearchCV(
        estimator=mp['model'], 
        param_grid=mp['params'], 
        cv=3,
        scoring='neg_mean_squared_error',
        n_jobs=-1
    )
    
    # Fit the model
    clf.fit(X_train, y_train)
    
    best_model = clf.best_estimator_
    y_pred = best_model.predict(X_test)
    
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    rho, p_value = spearmanr(y_test, y_pred)

    
    results.append({
        'Model': name,
        'Best Params': clf.best_params_,
        'Test RMSE': rmse,
        'Test R2 Score': r2,
        "Spearman's rho": rho,
        "p-value rho": p_value,
    })
    print(results[-1])

print("\nGrid Search Complete!\n")

results_df = pd.DataFrame(results)

results_df = results_df.sort_values(by='Test R2 Score', ascending=False).reset_index(drop=True)

print("=== Model Comparison ===")
print(results_df[['Model', 'Test RMSE', 'Test R2 Score', "Spearman's rho", "p-value rho"]])
print("\n=== Best Parameters ===")
for index, row in results_df.iterrows():
    print(f"{row['Model']}: {row['Best Params']}")
```

```python
import optuna
# from xgboost import XGBRegressor

def objective(trial):
    param = {
        'verbosity': 0,
        'objective': 'reg:squarederror',
        'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
        'max_depth': trial.suggest_int('max_depth', 2, 12),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.3, 1.0), # Vital para 358 variables
        'gamma': trial.suggest_float('gamma', 1e-8, 1.0, log=True),
        'reg_alpha': trial.suggest_float('reg_alpha', 1e-8, 10.0, log=True),
        'reg_lambda': trial.suggest_float('reg_lambda', 1e-8, 10.0, log=True),
        'random_state': 42
    }

    kf = KFold(n_splits=5, shuffle=True, random_state=42)

    model = XGBRegressor(**param)
    
    scores = cross_val_score(model, X_relev, y_reg, cv=kf, scoring='r2', n_jobs=-1)
    
    return scores.mean()

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=100)

print(f"Mejor R2 encontrado: {study.best_value:.4f}")
print("Mejores parámetros:", study.best_params)

best_model = XGBRegressor(**study.best_params)
best_model.fit(X_relev, y_reg)
```

## Encontrar deformaciones exactas para testear 

```python
def extend_feat(df_in, features):

    puntos_base = list(set([col.rsplit('_', 1)[0] for col in features if col.endswith(('_dx', '_dy', '_dz'))]))
    pares_puntos = list(combinations(puntos_base, 2))
    
    # 1. Creamos un diccionario para guardar temporalmente las nuevas columnas
    nuevas_columnas = {}
    
    for p1, p2 in pares_puntos:
        if all(f"{p}_{eje}" in df_in.columns for p in [p1, p2] for eje in ['dx', 'dy', 'dz']):
            nombre_col_dist = f"dist_{p1}_to_{p2}"
            nombre_col_angl = f"angl_{p1}_to_{p2}"
            
            dx_diff = df_in[f"{p1}_dx"] - df_in[f"{p2}_dx"]
            dy_diff = df_in[f"{p1}_dy"] - df_in[f"{p2}_dy"]
            dz_diff = df_in[f"{p1}_dz"] - df_in[f"{p2}_dz"]
    
            norm_p1 = np.sqrt(df_in[f"{p1}_dx"]**2 + df_in[f"{p1}_dy"]**2 + df_in[f"{p1}_dz"]**2)
            norm_p2 = np.sqrt(df_in[f"{p2}_dx"]**2 + df_in[f"{p2}_dy"]**2 + df_in[f"{p2}_dz"]**2)
            dot_p1_p2 = df_in[f"{p1}_dx"]*df_in[f"{p2}_dx"] + df_in[f"{p1}_dy"]*df_in[f"{p2}_dy"] + df_in[f"{p1}_dz"]*df_in[f"{p2}_dz"]
    
            # 2. Guardamos los resultados en el diccionario (súper rápido)
            nuevas_columnas[nombre_col_dist] = np.sqrt(dx_diff**2 + dy_diff**2 + dz_diff**2)
            # Nota: Agrupé la división en paréntesis por seguridad matemática
            nuevas_columnas[nombre_col_angl] = dot_p1_p2 / (norm_p1 * norm_p2)

    # 3. Convertimos el diccionario a un DataFrame de golpe
    df_nuevas = pd.DataFrame(nuevas_columnas)
    
    # 4. Concatenamos el original con las nuevas columnas (axis=1 es para unir columnas)
    df_extend = pd.concat([df_in.reset_index(drop=True), df_nuevas.reset_index(drop=True)], axis=1)

    return df_extend

extend_feat(df.iloc[[0]].copy(), features)
```

```python
nuevos_datos = []
num_relev_colums = len(relevant_features)
x_candidates = extend_feat(df_original_candidates[features], features)
num_colums = len(x_candidates.columns)

for row in x_candidates.itertuples(index=False):
    row_array = np.array(row)
    for i in range(len(features)*50):
        x_new_candidate = row_array + np.random.normal(0,0.001, num_colums)
        nuevos_datos.append(x_new_candidate)
        
df_aumentado = pd.DataFrame(nuevos_datos, columns=x_candidates.columns)
x_candidates = pd.concat([x_candidates, df_aumentado], ignore_index=True)

x_candidates['pred'] = best_model.predict(x_candidates[relevant_features]).flatten()
# x_good_candidates = x_candidates[x_candidates['pred'] < 0.95*x_candidates['pred'].min()]
# x_good_candidates = x_candidates[x_candidates['pred'] < -5.1]
x_good_candidates = x_candidates.copy()
x_good_candidates

# x_good_candidates[features+['pred']]
print(x_good_candidates['pred'].mean(), x_good_candidates['pred'].std(), x_good_candidates['pred'].min(), len(x_good_candidates['pred']) )
```
